# KERNEL DENSITY ESTIMATE PLOTS

```python
# STANDARTS
import numpy as np
import pandas as pd 
from numpy.random import randn

# STATS
from scipy import stats

# PLOTTING
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset, alpha = 0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)

x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

bandwidth = ( (4*dataset.std() ** 5) / (3 * len(dataset))) ** 0.2

kernel_list = []

for data_point in dataset:
    # Create a kernel for each point and append it to the kernel_list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    # Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4 
    
    plt.plot(x_axis, kernel, color = 'grey', alpha = 0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list, axis = 0)

fig = plt.plot(x_axis, sum_of_kde, color = 'indianred')

sns.rugplot(dataset)

plt.yticks([])

plt.suptitle("Sum of the basis function")
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color = 'black')

for bw in np.arange(0.5, 2, 0.25):
    sns.kdeplot(dataset, bw_method = bw, lw = 1.8, label = bw)
    
    
```

```python
kernel_options = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']

for kern in kernel_options:
    sns.kdeplot(dataset, label = kern)
```

```python
sns.kdeplot(y = dataset)
```

```python
sns.kdeplot(dataset, cumulative = True)
```

# HEATMAPS AND CLUSTERED MATRICES

```python
# STANDARTS
import numpy as np
import pandas as pd 
from numpy.random import randn

# STATS
from scipy import stats

# PLOTTING
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
flight_dframe = sns.load_dataset('flights')
```

```python
flight_dframe.head()
```

```python
flight_dframe = flight_dframe.pivot(index = 'year',columns = 'month',values = 'passengers')
```

```python
flight_dframe 
```

```python
sns.heatmap(flight_dframe )
```

```python
sns.heatmap(flight_dframe, annot = True, fmt = 'd')
```

```python
f,(axis1, axis2) = plt.subplots(2,1)

yearly_flights = flight_dframe.sum()

years = pd.Series(yearly_flights.index.values)
years = pd.DataFrame(years)

flights = pd.Series(yearly_flights.values)
flights = pd.DataFrame(flights)

year_dframe = pd.concat((years,flights),axis = 1)
year_dframe.columns = ['Year','Flights']

sns.barplot(x = 'Year', y = 'Flights', data = year_dframe, ax = axis1)

#sns.heatmap(flight_dframe, cmap = 'Blues', ax = axis2, cbar_kws{'orientation':'horizontal'})
```

```python
sns.clustermap(flight_dframe, standard_scale = 1)
```
